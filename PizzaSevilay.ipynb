```python
import csv
import datetime

#Menu.txt dosyası oluşturularak içine menüyü eklendi ve dosya kapatıldı.
metin_file = open('Menu.txt', "w")
metin_file.write('We got sauce selection input from user:\n1: Classic\n2: Margarita\n3: TürkPizza\n4: Sade Pizza\n\n* Please choose sauce:\n11: Olive\n12: Mushroom\n13: Goat cheese\n14: Meat\n15: Onion\n16: Corn\n* Thank You!')
metin_file.close()

class Pizza:
      def __init__(self, description, cost):
          self.description = description
          self.cost = cost
          

      def get_description(self):  
          return self.description

      def get_cost(self):
          return self.cost  

class Classic(Pizza):
      def __init__(self, description, cost):
          super().__init__(description, cost)
        
class Margarita(Pizza):
      def __init__(self, description, cost):
          super().__init__(description, cost)

class TurkPizza(Pizza):
      def __init__(self, description, cost):
          super().__init__(description, cost)

class SadePizza(Pizza):
      def __init__(self, description, cost):
          super().__init__(description, cost)      
    
class Extra:
      def __init__(self, description, cost):
          self.description = description
          self.cost = cost

      def get_description(self):  
          return self.description

      def get_cost(self):
          return self.cost

class Olive(Extra):
      def __init__(self, description, cost):
          super().__init__(description, cost) 
          self.description = description
          self.cost = cost 
class Mushroom(Extra):
      def __init__(self, description, cost):
          super().__init__(description, cost) 
          self.description = description
          self.cost = cost  

class Goatcheese(Extra):
      def __init__(self, description, cost):
          super().__init__(description, cost) 
          self.description = description
          self.cost = cost   

class Meat(Extra):
      def __init__(self, description, cost):
          super().__init__(description, cost) 
          self.description = description
          self.cost = cost   

class Onion(Extra):
      def __init__(self, description, cost):
          super().__init__(description, cost) 
          self.description = description
          self.cost = cost  

class Corn(Extra):
      def __init__(self, description, cost):
          super().__init__(description, cost) 
          self.description = description
          self.cost = cost           
       
class Decorator:
      def __init__(self, pizzaC, extraClasses):
          self.pizza = pizzaC
          self.extraDesc = ''
          self.extraTotal = 0
          for extraC in extraClasses :
              self.extraTotal += extraC.get_cost()
              self.extraDesc += ', extra' + extraC.get_description()

      def get_cost(self):
          return self.pizza.get_cost() + self.extraTotal

      def get_description(self):
          return self.pizza.get_description() + self.extraDesc

class Kullanici:
      def init(self, name, tc, credicard, password):
          self.name = name
          self.tc = tc
          self.credicard = credicard
          self.password = password      

      

def main():
    #We printed the txt file we created to the screen.
          with open('Menu.txt', "r") as metin_file:
              print(metin_file.read())

      #We received pizza selection input from user
          pizza_choice =int(input ("Pizza seçiminiz: "))
          if pizza_choice == 1:
              pizza = Classic("Classic Pizza", 50)
              print("It contains tomato onion cheese corn.")
          elif pizza_choice == 2:
              pizza = Margarita("Margarita Pizza", 45)
              print("It contains tomatoes and cheese.")
          elif pizza_choice == 3:
              pizza = Margarita("Turk Pizza", 60)
              print("It contains tomatoes, corn and meat.") 
          elif pizza_choice == 4:
              pizza = Margarita("Sade Pizza", 40)
              print("It contains cheese.")   
          else:
              print("you made the wrong choice")
              return None
         #We got sauce selection input from user.
          extra_list = []
          while True :
              extra_choice = input("Please choose your extra: (q for quit)")
              if extra_choice == 'q':
                  break
              elif extra_choice == '11': 
                  extra = Olive("Olive", 3)
                  extra_list.append(extra)
              elif extra_choice == '12': 
                  extra = Mushroom("Mushroom", 4)
                  extra_list.append(extra)
              elif extra_choice == '13': 
                  extra = Goatcheese("Goat cheese", 6)
                  extra_list.append(extra)
              elif extra_choice == '14': 
                  extra = Meat("Meat", 9)
                  extra_list.append(extra)
              elif extra_choice == '15': 
                  extra = Onion("Onion", 3)
                  extra_list.append(extra)
              elif extra_choice == '16': 
                  extra = Corn("Corn", 4)
                  extra_list.append(extra)
              else:
                  print("you made the wrong choice.")
                  return None

          pizza_with_extra = Decorator(pizza, extra_list)                      
      
      #user information input received.
          print("-------------------------------")
          print("Enter Your User Information")
          print("-------------------------------")
          name = input("Enter your name: ")
          tc = input("Enter your TC: ")
          credicard = input("Enter your credit card number:")
          password = input("Enter your Card Password:")

          print("-------------------------------")
          print("Your choice: ", pizza_with_extra.get_description() )
          print("Your total cost: ", pizza_with_extra.get_cost())

          now = datetime.datetime.now()
          with open('Orders_Database.csv', 'w', newline='') as writer_file:
              writer = csv.writer(writer_file) 
              writer.writerow(['name', 'tc', 'credi card nu.', 'password', 'order date', 'order choice', 'total cost'])
              writer.writerow([name, tc, credicard, password, now.strftime("%c"), pizza_with_extra.get_description(),pizza_with_extra.get_cost()  ])   
      
main()
    
       


```
